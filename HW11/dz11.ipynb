### Задача 2. Оценка

```python
df = pd.DataFrame()
p_mean, p_var = [], []
for i in range(15):
    r = poisson.rvs(17, size=100)
    p_mean.append(np.abs(np.mean(r) - 17))
    p_var.append(np.abs(np.var(r) - 17))
df['mean'] = p_mean
df['var'] = p_var
print(df)
print("Mean error for mean is", np.mean(df['mean']))
print("Mean error for variance is", np.mean(df['var']))
```

Точнее оценивает среднее арифметическое.

### Задача 3. 

```python
r = binom.rvs(5, 0.5, size=1000)
```

```python
m = np.mean(r)
# n * t
Var = np.var(r)
# n * t * (1-t)
t = -(Var/m-1)
print('m = ',  m/t, 'T = ', t)
```

### Задача 4.  

```python
df = pd.read_csv('breast_cancer_1000_genes.tsv', sep='\t', index_col = 'Unnamed: 0')
```

```python
df['max_min'] = df.max(axis=1) - df.min(axis=1)
df.sort_values('max_min', inplace = True)
df.drop(columns=['max_min'], inplace = True)
df.iloc[:10]
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=df.iloc[:10].T)
plt.show()
```

```python
df['var'] = df.var(axis=1)
df.sort_values('var', inplace = True)
df.drop(columns=['var'], inplace = True)
df.iloc[:10]
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=df.iloc[:10].T)
plt.show()
```

```python
df['qq'] = df.quantile(q=0.75, axis=1)-df.quantile(q=0.25, axis=1)
df.sort_values('qq', inplace = True)
df.drop(columns=['qq'], inplace = True)
df.iloc[:10]
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=df.iloc[:10].T)
plt.show()
```
