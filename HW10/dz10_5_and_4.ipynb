```python
from scipy.stats import t
from scipy.stats import norm
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Problem 4

```python
def draw_hist_pdf(a):
    rv1 = t(1)
    df = pd.DataFrame({})
        
    for i in range (1,10001):
        df[i] = rv1.rvs(size = a)

    df.loc["Mean"] = df.mean()

    df1 = pd.DataFrame({"x": np.linspace(-3, 3, 10000)})
    df1["PDF"] = rv1.pdf(df1["x"])
    plt.tight_layout()
        
    sns.histplot(x = df.loc["Mean"], stat = "density")
    sns.lineplot(data = df1, x = "x", y = "PDF", color = "Green")
    plt.xlim(-3, 3)
```

```python
draw_hist_pdf(2)
```

```python
draw_hist_pdf(5)
```

```python
draw_hist_pdf(14)
```

```python
draw_hist_pdf(100)
```

## Problem 5

```python
def draw_hist_pdf2(a):
    rv = t(3)
    df = pd.DataFrame({})

    for i in range(1,10001):
        df[i] = rv.rvs(size = a)
        
    df.loc["Mean"] = (df.mean() * np.sqrt(a)) / np.sqrt(3)

    df1 = pd.DataFrame({"x": np.linspace(-3, 3, 10000)})
    df1["PDF"] = norm(0, 1).pdf(df1["x"])
    plt.tight_layout()
        
    sns.histplot(x = df.loc["Mean"], stat = "density")
    sns.lineplot(data = df1, x = "x", y = "PDF", color = "Green")
    plt.xlim(-3, 3)
```

```python
draw_hist_pdf2(2)
```

```python
draw_hist_pdf2(5)
```

```python
draw_hist_pdf2(14)
```

```python
draw_hist_pdf2(100)
```
