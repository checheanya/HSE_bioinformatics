```python
import pandas as pd
```

```python
df = pd.read_csv('TCGA-COAD.tsv', sep='\t')
df.head(5)
```

## Задача 4

```python
genes_71 = set(df[['Gene', 'TCGA-A6-2671-11A']].sort_values(by='TCGA-A6-2671-11A').iloc[-100:]['Gene'].unique())
genes_75 = set(df[['Gene', 'TCGA-A6-2675-11A']].sort_values(by='TCGA-A6-2675-11A').iloc[-100:]['Gene'].unique())
```

```python
both = genes_71.intersection(genes_75)
len(list(both))
```

## Задача 5

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = df.set_index('Gene')
new = df.loc[list(both)][["TCGA-A6-2671-11A", "TCGA-A6-2675-11A"]]
new.shape
```

```python
new.reset_index(inplace=True)
```

```python
plt.figure(figsize=(15, 10))
sns.set(style='whitegrid', context='talk')
sns.scatterplot(x="TCGA-A6-2671-11A", y="TCGA-A6-2675-11A", data=new, hue='Gene')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
```

```python
fig, ax = plt.subplots(figsize=(17, 13))
ax.scatter(new["TCGA-A6-2671-11A"], new["TCGA-A6-2675-11A"])

for i, txt in enumerate(new['Gene']):
    ax.annotate(txt, (new["TCGA-A6-2671-11A"][i], new["TCGA-A6-2675-11A"][i]), arrowstyle['-'], textcoords="offset points", xytext=(0,10), ha='center')
```
